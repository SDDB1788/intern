```python
from bs4 import BeautifulSoup
import requests
```

```python
url=requests.get("https://www.imdb.com/list/ls091520106/")
url
```

```python
pg=url.content
```

```python
movie=BeautifulSoup(pg,'html.parser')
```

```python
movie.title.text
```

```python
movies=movie.findAll('div',{'class':'lister-item mode-detail'})
```

```python
Names=[]
Year=[]
Rating=[]

for i in movies:
    Names.append(i.find('h3',{'class':'lister-item-header'}).a.text)
    Year.append(i.find('span',{'class':'lister-item-year text-muted unbold'}).text[1:5])
    Rating.append(i.find('span',{'class':'ipl-rating-star__rating'}).text)
    
    
```

```python
data=list(zip(Names,Year,Rating))
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=['Names','Year','Rating'])
df.head()
```

# Mobile Phone under 20K-Amazon

```python
Url=requests.get("https://www.amazon.in/s?k=mobile+phone+under+20000&rh=n%3A976419031%2Cp_36%3A1318506031&dc&adgrpid=57838942246&ext_vrnc=hi&gclid=CjwKCAjwgOGCBhAlEiwA7FUXktkvBWiqNkXA6eCoMHHqKv41NSVfF9Qpt6olab5pf5xgLaVPpVYZhBoCLjcQAvD_BwE&hvadid=398120966129&hvdev=c&hvlocphy=9061994&hvnetw=g&hvqmt=e&hvrand=17360987953719657263&hvtargid=kwd-309190037721&hydadcr=26727_2177615&qid=1616422309&rnid=1318502031&tag=googinhydr1-21&ref=sr_nr_p_36_4")
Url
```

```python
Url.content
```

```python
Soup=BeautifulSoup(Url.content,'html.parser')
```

```python
titles=Soup.findAll('span',class_='a-size-base-plus a-color-base a-text-normal')
titles
```

```python
Mob=[]
for i in titles:
    Mob.append(i.text.replace('\n','')[0:17])

Mob
```

```python
prc=Soup.findAll('span',class_="a-price-whole")
prc
```

```python
Price=[]
for i in prc:
    Price.append(i.text.replace('\n',''))
Price
```

```python
rt=Soup.findAll('div',class_='a-row a-size-small')
rt
```

```python
Rating=[]
for i in rt:
    Rating.append(i.text.replace('\n',' ')[1:15])
Rating
```

```python
data=list(zip(Mob,Price,Rating))
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=['Mobile_Name','Price','Rating'])
df.head()
```
