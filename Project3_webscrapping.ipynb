# 1 & 2. Amazon

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.amazon.in/")
```

```python
search_tag=driver.find_element_by_xpath("//input[@id='twotabsearchtextbox']")
search_tag.send_keys('Guitar')
search_btn=driver.find_element_by_xpath("//input[@id='nav-search-submit-button']")
search_btn.click()
```

```python
Product_Nm=[]
Price=[]
DD=[]
Rating=[]

for r in range(0,3):
     for i in driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base a-text-normal']"):
        Product_Nm.append(i.text)
    for p in driver.find_elements_by_xpath("//span[@class='a-price-whole']"):
        Price.append(p.text)
    for d in driver.find_elements_by_xpath("//span[@class='a-text-bold']"):
        DD.append(d.text)
urls=[c.get_attribute("href")for c in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")]
for url in urls:
    driver.get(url)
    try:
        rate=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        rate.click()
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Rating.append(rating.text)
                
    except NoSuchElementException as e:
            Rating.append("No rating")
                
    
    
    
Rating[0:2]
Product_Nm[:2]    
    

    
```
