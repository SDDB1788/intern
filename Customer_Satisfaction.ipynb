```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
# importing libraries to train, test and predict the accuracy, as the target column is binary we have gone with logistic regression and other classifier models
# and the matrix to be accuracy_score,confusion_matrix,classification_report,GridSearchCV,cross_val_score
```

```python
lg=LogisticRegression()
```

```python
best_rstate=0
accu=0
for i in range(30,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    lg.fit(x_train,y_train)
    y_pred=lg.predict(x_test)
    tempaccu=accuracy_score(y_test,y_pred)
    if tempaccu>accu:
        accu=tempaccu
        best_rstate=i
        
print(f"Best accuracy {accu*100} found on random state {best_rstate}")
```

```python
#Best accuracy is found in random state 72 and splitting the model
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=72,stratify=y)
```

```python
print(x_train.shape,'\t\t',x_test.shape)
print(y_train.shape,'\t\t',y_test.shape)
```

```python
#checking the shape of train and test of target and other attributes
```

```python
parameters={'C':[1,10],'random_state':range(42,100)}
lg=LogisticRegression()
clf=GridSearchCV(lg,parameters)
clf.fit(x,y)
print(clf.best_params_)
```

```python
parameters={'n_neighbors':range(22,25)}
knn=KNeighborsClassifier()

clf=GridSearchCV(knn,parameters)
clf.fit(x,y)
print(clf.best_params_)
```

```python
# finding the best parameters for knn is 22 and Logistic is 42 (random state and C=1)
```

```python
LR=LogisticRegression(random_state=42)
KNN=KNeighborsClassifier(n_neighbors=22)
SVC=SVC()
```

```python
model=[LR,GaussianNB(),DecisionTreeClassifier(),KNN]

for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score of ',m,'is:')
    print(accuracy_score(y_test,predm)*100)
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('**************************************************')
    print('\n')
```

```python
#Decision tree gives the best accuracy, furthe checking through cross value score
```

```python
model=[LR,GaussianNB(),SVC,DecisionTreeClassifier(),KNN]
for m in model:
    score=cross_val_score(m,x,y,cv=10,scoring='accuracy')
    print('Model:',m)
    print('Score:',score*100)
    print('Mean Score:',score.mean())
    print('Standard Deviation:',score.std())
    print('**************************************************')
    print('\n')
```

```python
#Even here, the decision tree gives the best score of 88 percent
```
